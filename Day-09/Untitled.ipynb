# Colormaps
  - Classes of Colormaps
    - Sequential  - Used to represent information that has ordering. There is a change in lightness, often over a single hue.
    - Diverging   - Used when information plotted deviates around a middle value. Two colors are used in this case.
    - Qualitative - Used to represent information that does not have ordering or relationship (and is set of miscellaneous colors)

```python
odf.Sport.value_counts().plot(kind='pie', figsize=(11,11), colormap='Pastel1')
```

# GroupBy

```python
odf.groupby('Edition')
```

```python
type(odf.groupby('Edition'))
```

```python
list(odf.groupby('Edition'))
```

#### Iterating through group

```python
for group_key, group_value in odf.groupby('Edition'):
    print(group_key)
    print(group_value)
```

```python
type(group_key)
```

```python
type(group_value)
```

#### groupby functions
  - GroupBy.size()
  - GroupBy.count()
  - GroupBy.first(), GroupBy.last()
  - GrouBy.head(), GroupBy.tail()
  - GroupBy.mean()
  - Groupby.max(), GroupBy.min()

#### Aggregate
  - agg() Multiple Statistics on one calculation per group
  - DF.groupby(agg([...]))
  - DF.groupby(agg({ .. : [...] }))

```python
odf.groupby('Edition').size()
```

```python
# Aggregate
odf.groupby(['Edition', 'NOC', 'Medal']).agg(['min', 'max', 'count'])
```

```python
odf.groupby(['Edition', 'NOC', 'Medal']).agg({'Edition' : ['min', 'max', 'count']})
```

```python
odf.loc[odf.Athlete=='LEWIS, Carl'].groupby('Athlete'). agg({'Edition':['min', 'max', 'count']})
```
